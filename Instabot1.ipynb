```python
def like_first_30_posts():
    # just openinig the handler/profile of dilsefoodie........
    search=driver.find_element_by_class_name("XTCLo");
    search.send_keys("dilsefoodie");
    wait.until(EC.presence_of_element_located((By.CLASS_NAME,'yCE8d')))# this is for waiting for the dropdown to come..
    name=driver.find_element_by_class_name("yCE8d");# name is pointing to the handler of dilsefoodie now...
    name.click();
    # till now we have just opened the handler of dilsefoodie...it is similar to what we did for the dilsefoodie
    
    
    wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_9AhH0')));# this will wait until we don't find the posts...
    posts=driver.find_elements_by_class_name('_9AhH0');
    height_of_body=driver.execute_script('return document.body.scrollHeight');
    
    #suppose that net is slow and less posts are loaded then we will first load the 30 posts.......
    flag=False;
    if(len(posts)<30):
        #print("came_here");
        flaf=True;
        time.sleep(2);
        driver.execute_script('window.scrollBy(0,arguments[0])',height_of_body//5);
        posts=driver.find_elements_by_class_name('_9AhH0');
    
    # here we r liking the 1st 30 posts.......
    if(flag):
        driver.execute_script('window.scrollBY(0,-1000)');
   
    for post in posts[0:30]:
        post.click();
        wait.until(EC.presence_of_element_located((By.XPATH,"//button[@class='wpO6b ']")));
        time.sleep(1)
        wait.until(EC.element_to_be_clickable((By.XPATH,"//span[@class='fr66n']/button")));
        like_button=driver.find_element_by_xpath("//span[@class='fr66n']/button");
        driver.execute_script('arguments[0].click()',like_button);
        time.sleep(1)
        driver.back()# it doen't depend on the internet speed so.....we need to to just wait for the previous page.
        time.sleep(1)
    print("NOW I HAD LIKED THE 1ST 30 POSTS.......")
        
    time.sleep(3);#just giving a break of 3 seconds..between the liking and unliking of the posts.......
    
    
    for post in posts[0:30]:
        post.click();
        wait.until(EC.presence_of_element_located((By.XPATH,"//button[@class='wpO6b ']")));
        time.sleep(1);
        wait.until(EC.element_to_be_clickable((By.XPATH,"//span[@class='fr66n']/button")));
        like_button=driver.find_element_by_xpath("//span[@class='fr66n']/button");
        driver.execute_script('arguments[0].click()',like_button);
        time.sleep(1)
        driver.back()# it doen't depend on the internet speed so.....we need to to just wait for the previous page.
        time.sleep(1)
    print("NOW I HAD UNLIKED THE 1ST 30 POSTS.......")
```

```python
def users_i_follow():
    x_path="//div[@class='_47KiJ']/div[3]/a";
    driver.find_element_by_xpath(x_path).click();
    time.sleep(1);
    
    
    
    
    #getting the followers_list
    followers_link=driver.find_elements_by_class_name("-nal3 ");
    followers_link[1].click();
    # we need to wait for-sometime till when the pop-up appears.....
    wait.until(EC.presence_of_element_located((By.CLASS_NAME,'FPmhX')));
    # till now we have clicked and the pop-up have appeared.........
    index=0;
    previous_length=0;
    while True:
        followers_loaded=driver.find_elements_by_class_name('FPmhX');
        index=len(followers_loaded)-1;
        if(len(followers_loaded)>previous_length):
            break;
        else:
           # print(followers_loaded[index].text)
            previous_length=len(followers_loaded);
            followers_loaded[index].send_keys(Keys.END);
            time.sleep(1);
    my_followers=[i.text for i in followers_loaded];
    driver.back();
    
    
    
    #getting_list_of_persons_i_follow...i'e..following..
    time.sleep(1);
    following_link=driver.find_elements_by_class_name("-nal3 ");
    following_link[2].click();
    # we need to wait for-sometime till when the pop-up appears.....
    wait.until(EC.presence_of_element_located((By.CLASS_NAME,'FPmhX')));
    # till now we have clicked and the pop-up have appeared.........
    index=0;
    previous_length=0;
    while True:
        followings_loaded=driver.find_elements_by_class_name('FPmhX');
        index=len(followings_loaded)-1;
        if(len(followings_loaded)>previous_length):
            break;
        else:
            #print(followings_loaded[index].text)
            previous_length=len(followings_loaded);
            followings_loaded[index].send_keys(Keys.END);
            time.sleep(1);
    my_followings=[i.text for i in followings_loaded];
    driver.back();
    
    #print(my_followers);
    #print(my_followings);
    followers_that_dont_follow_me=[];
    for i in my_followers:
        if i not in my_followings:
            followers_that_dont_follow_me.append(i);
    
    return (followers_that_dont_follow_me);        
```
